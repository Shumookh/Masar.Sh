**Cleaning & Visualation Data**

```python
# Cleaning Data
      # 1- import library & Data (from kagle)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/global-data-on-sustainable-energy (1).csv')
df
```

```python
# get the column
df.columns
```

```python
#check how many missing values
df.isnull().sum()
```

```python
# dropping the missing values & Check of null
df_new = df.dropna()
df_new.isnull().sum()
```

```python
df.info()
```

```python
df_new.info()
```

**Data visualization**

```python
# 1- Line plot for Electricity from renewables (TWh) over Years
# Set up the matplotlib figure
plt.plot(df_new['Year'], df_new['Electricity from renewables (TWh)'])
    #label the x,y axis
plt.xlabel('Year')
plt.ylabel('Electricity from renewables (TWh)')
plt.title('Electricity from renewables (TWh) Over Years (Line Chart)')
plt.grid(True)
    # Show the plot
plt.show()
```

```python
# 2-Bar Chart for Electricity from renewables (TWh) over Years
plt.bar(df_new['Year'], df_new['Electricity from renewables (TWh)'])
   #label the x,y axis
plt.xlabel('Year')
plt.ylabel('Electricity from renewables (TWh)')
plt.title('Electricity from renewables (TWh) Over Years (Line Chart)')
plt.grid(True)
   # Show the plot
plt.show()
```

```python
# 3- Bar plot for Electricity from renewables (TWh) over Years
sns.barplot(x="Year", y= "Electricity from renewables (TWh)", data= df_new, palette="YlGnBu")
```

```python
# 4- Heat map Plot for Sales, Profit, and Expenses over Years
#split important data to used
df_X = df_new['Electricity from renewables (TWh)'], df_new['Year']
sns.heatmap(df_X)
```

```python
# 5- scatter Chart for Electricity from renewables (TWh) over Years
x = df_new['Year']
y = df_new['Electricity from renewables (TWh)']
    #label the x,y axis
plt.xlabel('Year')
plt.ylabel('Electricity from renewables (TWh)')
plt.title('Electricity from renewables (TWh) Over Years (scatter Chart)')
plt.grid(True)
plt.scatter(x, y)
plt.show()
```
